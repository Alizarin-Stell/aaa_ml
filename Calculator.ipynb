```python
import numpy as np

x = np.array([
    [0, 1, 3, 4, 0],
    [0, 3, -1, 1, 0]])

w = np.array([
    [2, 1, -1],
    [-1, 2, 3]])

y1 = w * x[:, :3]
y2 = w * x[:, 1:4]
y3 = w * x[:, 2:5]
(y1 + y2 + y3).sum()
```

```python
x_1 = np.array([[7], [-4]])
x_2 = np.array([[0.5], [3]])
x_3 = np.array([[10], [2]])

W_aa = np.array([
                [0, 1, 0],
                [2, 0, -1],
                [0, 1, 3]])
W_ax = np.array([
                [0, 2],
                [1, 0],
                [0, 4]])
b_a = np.array([[-1], [1], [0]])

def relu(x):
  return np.where(x > 0, x, 0)
```

```python
y1 = W_ax @ x_1 + b_a
y2 = W_aa @ relu(y1) + W_ax @ x_2 + b_a
y2
```
