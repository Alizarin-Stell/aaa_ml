```python
import pandas as pd
```

```python
df1 = pd.read_csv(way + 'toloka_ex.tsv', sep='\t')
df2 = pd.read_csv(way + 'toloka_ed.tsv', sep='\t')
df1
```

```python
df = pd.read_csv(way + 'toloka.tsv', sep='\t')
df = df.dropna(subset=['INPUT:image'])
df.head(5)
```

```python
combined_df = pd.concat([df, df1, df2], ignore_index=True)
```

```python
def check_all_equal(x):
    return x.nunique() == 1

combined_df.groupby("INPUT:image")["OUTPUT:result"].agg(check_all_equal)
```

```python
combined_df['image_id_ext'] = combined_df['INPUT:image'].apply(lambda x: x.split('/')[-1].split('.')[0])
combined_df['result'] = combined_df.apply(lambda x: x['GOLDEN:result'] if pd.notnull(x['GOLDEN:result']) else x['OUTPUT:result'], axis=1)
combined_df['result'] = combined_df['result'].astype(int)
final_df = combined_df[['image_id_ext', 'result']]
final_df_unique = final_df.drop_duplicates(subset=['image_id_ext'], keep='first')
final_df_unique
```

```python
final_df_unique.to_csv('final_dataset.csv', index=False)
```
